```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('vehicles.csv') # lendo os dados

car_data.head(10)
```

```python
year_price = car_data.groupby('model_year')['price'].value_counts().reset_index()


px.bar(year_price,x='model_year',y='price',color='price')
```

```python
condition_price = car_data.groupby('condition')['price'].value_counts().reset_index()

px.scatter(condition_price,y='condition',x='price',color='price')
```

```python
fuel_price = car_data.groupby('fuel')['price'].value_counts().reset_index()

px.box(fuel_price,x='fuel',y='price',color='fuel')
```

```python
odometer_price = car_data.groupby('odometer')['price'].value_counts().reset_index()

px.scatter(odometer_price,y='price',x='odometer',color='odometer')
```

```python
type_price = car_data.groupby('type')['price'].value_counts().reset_index()
px.scatter(type_price,y='type',x='price',color='type')
```
